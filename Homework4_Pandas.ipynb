```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["$0-585", "$585-615", "$615-645", "$645-675"]

perspending_df=pd.DataFrame(school_data_complete)
perspending_budget=perspending_df['budget']
perspending_size=perspending_df['size']
perspending_perstudent=perspending_budget/perspending_size
```

```python

perspending_df['Spending Per Student']=perspending_perstudent

perspending_df['Spending Ranges'] = pd.cut(perspending_df['Spending Per Student'], spending_bins, labels=group_names)
```

```python

pass_math_loc_df= perspending_df.loc[perspending_df['math_score']>69].groupby(['Spending Ranges'])
mathpass_counts=pass_math_loc_df.math_score.count()
mathpass_counts
```

```python
counts_perbin=perspending_df["Spending Ranges"].value_counts()
counts_perbin
```

```python
pass_reading_loc_df=perspending_df.loc[perspending_df['reading_score']>69].groupby(['Spending Ranges'])
readpass_counts=pass_reading_loc_df.reading_score.count()
readpass_counts
```

```python
perspending_math_pass=(mathpass_counts/counts_perbin)*100
perspending_math_pass
```

```python



pass_math_loc_df= perspending_df.loc[perspending_df['math_score']>69].groupby(['Spending Ranges'])

pass_reading_loc_df=perspending_df.loc[perspending_df['reading_score']>69].groupby(['Spending Ranges'])

passedperspending_df=perspending_df.groupby('Spending Ranges')

passed_math=passedperspending_df.math_score.mean()

passed_reading=passedperspending_df.reading_score.mean()

counts_perbin=perspending_df["Spending Ranges"].value_counts()
perspending_math_pass=(mathpass_counts/counts_perbin)*100
perspending_read_pass=(readpass_counts/counts_perbin)*100
overall_pass=(perspending_math_pass+perspending_read_pass)/2


school_by_spending_summary=pd.DataFrame({
    'Average Math Score': passed_math, 
    'Average Reading Score': passed_reading,  
    '% Passing Math': perspending_math_pass, 
    '% Passing Reading': perspending_read_pass, 
    '% Overall Passing Rate': overall_pass
})
school_by_spending_summary





```

```python
size_bins = [0, 1000, 2000, 5000]
size_group_names = ["Small (0-1000)", "Medium (1000-2000)", "Large (2000-5000)"]
size_df=school_data_complete
size_df["Size Ranges Per Students"]=pd.cut(size_df['size'],size_bins,labels=size_group_names)

math_scores_ss=size_df.loc[size_df['math_score']>69].groupby(['Size Ranges Per Students'])
reading_scores_ss=size_df.loc[size_df['reading_score']>69].groupby(['Size Ranges Per Students'])
school_size=size_df.groupby('Size Ranges Per Students')
ss_math_mean=school_size.math_score.mean()
ss_reading_mean=school_size.reading_score.mean()
bin_size_count=size_df["Size Ranges Per Students"].value_counts()
ss_math_pass=(math_scores_ss.math_score.count()/bin_size_count)*100
ss_read_pass=(reading_scores_ss.reading_score.count()/bin_size_count)*100
overall_pass_rate=(ss_math_pass+ss_read_pass)/2
school_size_frame=pd.DataFrame({
    'Average Math Score': ss_math_mean, 
    'Average Reading Score': ss_reading_mean,  
    '% Passing Math': ss_math_pass, 
    '% Passing Reading': ss_read_pass, 
    '% Overall Passing': overall_pass_rate
})
school_size_frame
```
